# Import

```python
import os
import shutil

import pandas as pd
```

```python
path = "/workspace/Storage/ninja-turtles/Data/pred_OD_baseline_test"
paths = [os.path.join(path, x) for x in os.listdir(path)]
```

```python
columns = ['label', 'x', 'y', 'w', 'h', 'confidence']
df = pd.DataFrame()
for p in paths:
    with open(p, 'r') as f:
        d = list(map(lambda x: x.strip().split(' '), f.readlines()))
    df_ = pd.DataFrame(d, columns = columns)
    df_['path'] = p
    df = pd.concat([df, df_])
df = df.reset_index(drop=True)
```

# analysis

분포 확인

```python
df.groupby(by='label').count()
```

가장 적은 라벨 수(152) 만큼 샘플링

```python
df_sampled = df.groupby('label').apply(lambda x: x.sample(n=152, random_state=42)).reset_index(drop=True)
```

```python
df_sampled.groupby(by='label').count()
```

총 라벨링 파일 수

```python
len(df_sampled['path'].unique())
```

라벨링 대상 이미지 세트 생성

```python

for pp in df_sampled['path'].unique():
    pp = pp.replace('.txt','.jpg').replace("pred_OD_baseline_test","test/images")
    tp = pp.replace("test","labeling_OD")
    shutil.copy(pp,tp)
    # print(pp,tp)
```
