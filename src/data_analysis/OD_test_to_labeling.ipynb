```python
class Node:
    def __init__(self, path, values):
        self.path = path
        self.values = values

    def __hash__(self):
        return hash(self.path)

    def __eq__(self, other):
        return self.path == other.path
    
class Group:
    def __init__(self, nodes = []):
        self.paths = []
        self.values = {}
        self.state = True
        for n in nodes:
            self.add(n)
            if self.state == False:
                break

    def add(self, n):
        self.paths.append(n.path)
        for v in n.values:
            if self.values.get(v, 0):
                self.values[v] += 1
            else:
                self.values[v] = 1
            if self.values[v]>10:
                self.state = False
                break
```

```python
import copy
import pdb
def get_item(nodes, root):
    # pdb.set_trace()
    result = []
    for node in nodes:
        root_ = copy.deepcopy(root)
        nodes_ = copy.deepcopy(nodes)
        nodes_.pop(nodes.index(node))
        root_.add(node)
        if root_.state:
            result.append(root_)
            roots = get_item(nodes_, root_)
            result+=roots
    return result
```
